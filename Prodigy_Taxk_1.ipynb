"Create a bar chart or histogram to visualize the distribution of a categorical or continuous variable, such as the distribution of ages or genders in a population."

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data=pd.read_csv('world_population_data.csv')
```

```python
print(data.to_string())
```

```python
print(data.head())
```

```python
print(data.tail())
```

```python
print(data.shape)
```

```python
country_name="United States"
country_data=data[data['country']==country_name]
years=country_data.columns[4:13]
population=country_data.iloc[ : ,4:13].values.flatten()
plt.figure(figsize=(20,10))
plt.bar(years,population,color='red')
plt.xlabel('Years')
plt.ylabel("Population")
plt.title(f"Population of {country_name} Over Various Years")
plt.show()
```

```python
plt.hist(population,bins=10,color='blue',edgecolor='black')
plt.show()
```
