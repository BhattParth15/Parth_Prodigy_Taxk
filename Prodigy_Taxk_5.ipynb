#### Analyze traffic accident data to identify patterns related to road conditions, weather, and time of day. Visualize accident hotspots and  contributing factors.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('accidents.csv')
df.head()
```

```python
## Dropping unnecessary columns
df.drop(columns=['Accident_Index','Number_of_Vehicles','Number_of_Casualties','Weather_conditions'],axis=1,inplace=True)
df.info()

```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
#Distribution of accidents across different Road_conditions
plt.figure(figsize=(10,5))
sns.countplot(x="Road_conditions",data=df)
plt.title("Distribution of accidents across different Road_conditions")
plt.show()
```

```python
#For Road_type
sns.countplot(y="Road_type",data=df)
plt.title("Accidents for Road type")
plt.show()
```

```python
#For Light_conditions
sns.countplot(x="Light_conditions",data=df)
plt.title("Accidents for Light_conditions")
plt.show()
```

```python
#For day timing
sns.countplot(x="Time",data=df)
plt.title("Accidents for Time")
plt.show()
```

```python
#For speed limit
sns.countplot(x="Speed_limit",data=df)
plt.title("Accidents for Speed_limit")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='Road_conditions',hue='Road_type',data=df)
plt.title("Accidents by road conditions and Road type")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='Speed_limit',hue='Light_conditions',data=df)
plt.title("Accidents by Speed limit and Light_conditions")
plt.show()
```

#### Light conditions is Daylight and speed is high that time most of accidents occur.
