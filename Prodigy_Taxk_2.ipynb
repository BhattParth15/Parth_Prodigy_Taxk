# Checking null or missing value

```python
print(data.isnull().sum())
```

# Droping unnecessary columns

```python
print(data.drop(columns="Cabin",axis=1,inplace=True))
```

# Filling the missing values of the age column by Mean value of the column

```python
data.loc[:,'Age']=data['Age'].fillna(data['Age'].mean())
```

# Filling the messing values of the Fare column by Mean value of the column

```python
data.loc[:,'Fare']=data['Fare'].fillna(data['Fare'].mean())
```

# Distribution of Fare and age

# Survival rate in different age category

```python
#Age cutoff point and it's label
cutoff_point=[0,5,12,18,35,60,100]
label_names=['Infant','Child','Teenager','Young Adult','Adult','Senior']
#Create the "Age_categories"column
data['Age_categories']=pd.cut(data['Age'],bins=cutoff_point,labels=label_names,right=False)
#Creating pivot table for survival rates based on age categories
age_pivot=data.pivot_table(index='Age_categories',values='Survived',observed=False)
#Define colors for each bar
colors=['blue','green','orange','purple','pink','black']
#Plotting the barchart with different colors
fig,ax=plt.subplots()
bars=ax.bar(age_pivot.index,age_pivot['Survived'],color=colors)
handles=[plt.Rectangle((0,0),1,1,color=colors[i]) for i in range(len(colors))]
ax.legend=(handles,label_names)
ax.set_title=("Survival rate by age categories")
ax.set_xlabel=("Age Category")
ax.set_ylabel=("Survival Rate")
plt.xticks(rotation=40)
plt.show()
```
